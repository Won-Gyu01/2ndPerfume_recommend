```python
import pandas as pd
import os
import shutil

# 경로 설정
base_folder = "C:/PerfumeS/spm/NEW_FACE_DATAFOLDER"  # 감정 폴더들이 위치한 기본 폴더
target_folder = "C:/PerfumeS/spm/NEW_FACE_DATA"  # 모든 이미지를 저장할 하나의 폴더
os.makedirs(target_folder, exist_ok=True)

# CSV 파일 불러오기
csv_path = "C:/PerfumeS/spm/NEW_FACE_DATAFOLDER/labels.csv"
updated_csv_path = "C:/PerfumeS/spm/NEW_FACE_DATA/updated_labels.csv"
labels_df = pd.read_csv(csv_path)

# 새 CSV 파일을 위한 리스트
new_labels = []

# 폴더 내의 모든 감정별 이미지 처리
for emotion_folder in os.listdir(base_folder):
    emotion_path = os.path.join(base_folder, emotion_folder)
    
    # 디렉터리인지 확인
    if not os.path.isdir(emotion_path):
        continue  # 디렉터리가 아니면 건너뛰기
    
    # 폴더 내 이미지들 복사
    for image_name in os.listdir(emotion_path):
        original_image_path = os.path.join(emotion_path, image_name)
        
        # 고유 이름 생성 (ex: happy_image0001.jpg)
        unique_image_name = f"{emotion_folder}_{image_name}"
        new_image_path = os.path.join(target_folder, unique_image_name)
        
        # 이미지 복사 및 라벨 정보 업데이트
        shutil.copy(original_image_path, new_image_path)

        # CSV에서 해당 이미지 이름을 가진 행 찾기 및 업데이트
        label_row = labels_df[labels_df["pth"].str.contains(image_name)]
        
        # 새 라벨 정보 추가
        if not label_row.empty:
            new_labels.append({
                "image_name": unique_image_name,
                "label": label_row.iloc[0]["label"],
                "relFCs": label_row.iloc[0]["relFCs"]
            })

# 새로운 CSV 파일로 저장
new_labels_df = pd.DataFrame(new_labels)
new_labels_df.to_csv(updated_csv_path, index=False)

print("모든 이미지와 라벨 파일이 고유한 이름으로 업데이트되었습니다.")
```

```python
import pandas as pd
import os

# updated_labels.csv 파일과 라벨 파일 저장 폴더 경로 설정
csv_path = "C:/PerfumeS/spm/NEW_FACE_DATA/filtered_labels.csv"
label_folder = "C:/PerfumeS/spm/NEW_FACE_DATA/labels"  # 각 이미지의 라벨 파일을 저장할 폴더
os.makedirs(label_folder, exist_ok=True)

# CSV 파일 로드
labels_df = pd.read_csv(csv_path)

# CSV 파일의 라벨 목록에 따라 class_mapping 생성
unique_labels = labels_df["label"].unique()
class_mapping = {label: idx for idx, label in enumerate(unique_labels)}
print("Updated Class Mapping:", class_mapping)

# 라벨 파일 생성
for index, row in labels_df.iterrows():
    image_name = row["image_name"]
    label = row["label"]
    class_id = class_mapping.get(label, -1)
    
    if class_id == -1:
        print(f"Skipping label '{label}' for image '{image_name}' - Not found in class_mapping")
        continue
    
    # 라벨 파일 경로
    label_file_path = os.path.join(label_folder, image_name.replace(".jpg", ".txt"))
    yolo_format = f"{class_id} 0.5 0.5 1.0 1.0\n"
    
    # 파일 생성
    try:
        with open(label_file_path, "w") as f:
            f.write(yolo_format)
        print(f"File saved: {label_file_path}")
    except Exception as e:
        print(f"Error writing file {label_file_path}: {e}")

print("CSV 파일에서 개별 라벨 파일로 변환 완료.")
```

# 이미지 학습

```python
import os
import shutil
import random

# 이미지와 라벨 파일이 있는 폴더 경로 설정
image_folder = "C:/PerfumeS/spm/NEW_FACE_DATA"
label_folder = "C:/PerfumeS/spm/NEW_FACE_DATA/labels"

# 분할된 데이터셋을 저장할 폴더 경로 설정
output_folder = "C:/PerfumeS/spm/NEW_FACE_TRAIN/DataSet"
train_img_folder = os.path.join(output_folder, "images/train")
val_img_folder = os.path.join(output_folder, "images/val")
test_img_folder = os.path.join(output_folder, "images/test")
train_label_folder = os.path.join(output_folder, "labels/train")
val_label_folder = os.path.join(output_folder, "labels/val")
test_label_folder = os.path.join(output_folder, "labels/test")

# 폴더 생성
os.makedirs(train_img_folder, exist_ok=True)
os.makedirs(val_img_folder, exist_ok=True)
os.makedirs(test_img_folder, exist_ok=True)
os.makedirs(train_label_folder, exist_ok=True)
os.makedirs(val_label_folder, exist_ok=True)
os.makedirs(test_label_folder, exist_ok=True)

# 이미지 파일 목록 불러오기
image_files = [f for f in os.listdir(image_folder) if f.endswith('.jpg')]

# 데이터를 셔플하고 분할
random.shuffle(image_files)
train_split = int(len(image_files) * 0.8)
val_split = int(len(image_files) * 0.9)

train_files = image_files[:train_split]
val_files = image_files[train_split:val_split]
test_files = image_files[val_split:]

# 분할된 파일을 각각의 폴더로 복사
for file_list, img_folder, lbl_folder in [(train_files, train_img_folder, train_label_folder),
                                          (val_files, val_img_folder, val_label_folder),
                                          (test_files, test_img_folder, test_label_folder)]:
    for file_name in file_list:
        # 이미지와 라벨 파일 경로
        img_src = os.path.join(image_folder, file_name)
        lbl_src = os.path.join(label_folder, file_name.replace(".jpg", ".txt"))
        
        # 이미지와 라벨 파일 복사
        shutil.copy(img_src, os.path.join(img_folder, file_name))
        shutil.copy(lbl_src, os.path.join(lbl_folder, file_name.replace(".jpg", ".txt")))

print("데이터셋이 train, val, test로 나누어졌습니다.")
```
